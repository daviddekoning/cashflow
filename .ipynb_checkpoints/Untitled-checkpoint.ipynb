```python
from datetime import date, timedelta
from sys import stdout

from cashflow import CompositeCashflow, IntervalCashflow, MonthlyCashflow, StartOn, OneTimeCashflow, Limited
from cashflow import runCashflows, sumCashflows

from plotly.offline import plot
import plotly.graph_objs as go

startDate = date( 2018,1,14 )
endDate = date(2018,12,30)
length = (endDate - startDate).days
starting_balance = 2000
cashflows = []

pay = CompositeCashflow("Arup pay")
pay.add( IntervalCashflow( "Arup Pay", date( 2018, 1, 12), 14, 2889.19 ) )

pay.add( Limited( IntervalCashflow("EI", date( 2018, 1, 12), 14, -58.3), date(2018, 1, 1 ), 672.10) )
pay.add( Limited( IntervalCashflow("QPIP", date( 2018, 1, 12), 14, -24.58), date(2018, 1, 1 ), 405.52) )
pay.add( Limited( IntervalCashflow("QPP", date( 2018, 1, 12), 14, -264.07), date(2018, 1, 1 ), 2829.60) )

pay.add( MonthlyCashflow( "Arup Profit Share", 19, 2500, months=[6,11]) )
pay.add( MonthlyCashflow( "Tax Return", 15, 3000, months=[4] ) )
cashflows.append( pay )

childBenefit = CompositeCashflow("Child Benefit")
childBenefit.add( MonthlyCashflow( "Child Benefit", 20, 583.54) )
childBenefit.add( OneTimeCashflow( "Quebec Benefit Backlog", date(2018,4,3), 160*8 ) )
childBenefit.add( MonthlyCashflow( "Quebec benefit", 3, 360, months=[1,4,7,10]) )
cashflows.append( childBenefit )

cashflows.append( MonthlyCashflow( "RESP", 1, -120) )

giving = CompositeCashflow("Giving")
giving.add( IntervalCashflow( "Tithe", date(2018, 1, 12+5), 14, -288.92 ) )
giving.add( Limited( IntervalCashflow( "Tithe", date(2018, 1, 12+5), 14, 58.3 ), date(2018, 1, 1 ), 67.21 ) )
giving.add( Limited( IntervalCashflow( "Tithe", date(2018, 1, 12+5), 14, 2.49 ), date(2018, 1, 1 ), 40.55 ) )
giving.add( Limited( IntervalCashflow( "Tithe", date(2018, 1, 12+5), 14, 26.41 ), date(2018, 1, 1 ), 282.96 ) )
giving.add( MonthlyCashflow( "Nico & Molly", 11, -350) )
cashflows.append( giving )

cashflows.append( IntervalCashflow( "Mortgage", date( 2016, 8, 15), 14, -523.97 ) )
cashflows.append( MonthlyCashflow( "Maintenance", 1, -300 ) )
cashflows.append( MonthlyCashflow( "Hydro", 2, -285, months=[2,4,12]) )
cashflows.append( MonthlyCashflow( "Hydro", 2, -100, months=[6,8,10]) )

condo = CompositeCashflow("Condo Costs")
condo.add( MonthlyCashflow( "Bank fees", 25, -25, months =[10]))
condo.add( MonthlyCashflow( "School Tax", 20, -117.27, months=[7,10] ) )
condo.add( MonthlyCashflow( "Municipal Tax", 15, -503.65, months=[2,5] ) )
condo.add( MonthlyCashflow( "Condo Fees", 10, -103.32-68.56) )
condo.add( OneTimeCashflow( "Condo Fees", date(2018,1,31), -113) )
cashflows.append( condo )

cashflows.append( MonthlyCashflow( "Insurance", 1, -58.31 ) )
cashflows.append( MonthlyCashflow( "Teksavvy", 7, -42) )
cashflows.append( MonthlyCashflow( "Koodo", 23, -40.78) )

daycare = CompositeCashflow("Day Care")
daycare.add( IntervalCashflow( "Day care", date(2017,9,12), 7, -40, date(2018,6,15) ) )
cashflows.append( daycare )

cashflows.append( IntervalCashflow( "E's", date( 2016,1,29 ), 7, -260) )
cashflows.append( IntervalCashflow( "D's", date( 2016,1,29), 7, -40) )
cashflows.append( IntervalCashflow( "Service de Garde", date( 2017,9,11), 7, -50) )
cashflows.append( MonthlyCashflow( "STM pass", 1, -83, months=[1,2,3,4,11,12] ) )
cashflows.append( MonthlyCashflow( "Clothing", 1, -150 ) )

counselling = CompositeCashflow("Counselling")
counselling.add( MonthlyCashflow( "Iris", 12,-76.5) )
counselling.add( IntervalCashflow("Gregor", date(2016, 11, 1), 28, -100) )
counselling.add( OneTimeCashflow( "Reimbursement", date(2018, 5, 20 ), 500 ) )
cashflows.append( counselling )

vacation = CompositeCashflow("Vacation")
vacation.add( OneTimeCashflow("Flights", date(2018,2,12), -972*2))
vacation.add( OneTimeCashflow("Vacation costs", date(2018,4,16), -1500))
cashflows.append( vacation )

misc = CompositeCashflow("Misc.")
misc.add( MonthlyCashflow( "Apple Music", 5, -14.99) )
misc.add( MonthlyCashflow( "Netflix", 11, -8 ) )
misc.add( MonthlyCashflow( "Audible", 1, -21) )
misc.add( OneTimeCashflow( "Swimming Lessons", date(2017,8,30), -78 * 2) )
misc.add( OneTimeCashflow( "Swimming Lessons", date(2018,1,1), -78 * 2) )
misc.add( OneTimeCashflow( "Swimming Lessons", date(2017,3,1), -78 * 2) )
misc.add( OneTimeCashflow( "Dentist cash", date(2018,1,29), 386 ) )
cashflows.append( misc )
```
